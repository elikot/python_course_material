## Fitting a function to data

In this exercise we will:

- Generate data to fit
- Fit a function to this data
- Plot the function and data together

### Reading in data

```python
import pandas

filename = 'data/example_data.txt'
df = pandas.read_csv(filename, sep=',\s*') # koma separator, get rid of spaces *
print(df.keys())
```

Now we can extract the individual replicates and normalize them:

```python
import numpy 

max_value = numpy.array([df['A'].max(), 
                         df['B'].max(), 
                         df['C'].max(),
                        ], dtype=numpy.float64).max() # all converted to data array fype float 64, leiame nendest 3s suurima

nrA = df['A'] / max_value
nrB = df['B'] / max_value
nrC = df['C'] / max_value

replicates = pandas.DataFrame([nrA, nrB, nrC])

assert df['x'].shape == nrA.shape #Shape should be the same : if its not true, stop (shape - rows, columns)
```

### Plotting data

```python
import math
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6,4))

plt.plot(df['x'], nrA, 'r.', label='A')
plt.plot(df['x'], nrB, 'b.', label='B')
plt.plot(df['x'], nrC, 'g.', label='C')

plt.errorbar(df['x'], replicates.mean(), fmt='k.', alpha=0.2,        # fmt=k. , alpha 0.2- transparent ,black
             yerr=math.sqrt(2)*replicates.std(), label='mean with error') # yerr- y direction error,

plt.legend(loc='best')
plt.xlabel('Concentration')
plt.ylabel('Measured data')
plt.title('Concentration measurements')
plt.xlim(0, 6.5)
show(fig)
```

### Pick a fitting function

It looks like sigmoidal data.  Perhaps we should use a sigmoical function:

$$ f(x) = \frac{1}{1+e^{-k \cdot (x-x_0)}}$$

where $x$ is the concentration, $x_0$ is the offset in concentration, and $k$ is the exponential factor.  The unknown parameters are $x_0$ and $k$.  

This function written in Python

```python
def fit_function(x, x0, k):  # x0 - moving around x ais
     return 1 / (1 + numpy.exp(-k*(x-x0)))
```

### Finding the unknown parameters using least squares fitting

```python
from scipy.optimize import curve_fit  # curve fit- 

# We include all data in the analysis:
all_x_values = numpy.concatenate([df['x']]*3) # all of the data together
all_measured_values = numpy.concatenate((nrA, nrB, nrC)) # all the measured values together

optimal_parameters, cov_mat = curve_fit(fit_function, 
                                        all_x_values, 
                                        all_measured_values)
```

```python
print('''Optimal parameters:\nx0: {1:.2f}\n k: {2:.2f}\n
The diagonals provide the variance of the parameter estimate:
{0}'''.format(cov_mat, *optimal_parameters))
```

```python
x_range = numpy.linspace(0, df['x'][0], 100) # tee x valued , linespace   0 to the largest , 100 points
estimated_y = fit_function(x_range, *optimal_parameters) # *optimal_parameters - duple lahtipakkimine  nr 1 ja 2 sealt: {1: } {2: ]
```

```python
import math
fig = plt.figure(figsize=(6,4))

plt.plot(df['x'], nrA, 'r.', label='A')
plt.plot(df['x'], nrB, 'b.', label='B')
plt.plot(df['x'], nrC, 'g.', label='C')

plt.errorbar(df['x'], replicates.mean(), fmt='k.', alpha=0.2, 
             yerr=math.sqrt(2)*replicates.std(), label='mean with error')

plt.plot(x_range, estimated_y, 'k-', alpha=0.6, label='fit')

plt.legend(loc='best')
plt.xlabel('Concentration',color='red',fontsize=18)
plt.ylabel('Measured data',color='green',fontsize=18)
plt.title('Concentration measurements')

plt.xlim(0, 6.5)
show(fig)
```
