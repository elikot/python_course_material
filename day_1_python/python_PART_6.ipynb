## Exceptions

In Python errors are managed with a special language construct called "Exceptions". When errors occur exceptions can be raised. This interrupts program flow and directs the Python interpreter to the nearest place where the raised exception can be caught.  If the exception 
is not caught, the program terminates.


To generate an exception we can use the `raise` statement, which takes an argument that must be an instance of the class `BaseExpection` or a class dervied from it. 

```python
additional_object = dict(param='bad')
raise Exception("description of the error", additional_object)
```

A typical use of exceptions is to abort functions when some error condition occurs, for example:

    def my_function(arguments):
    
        if not verify(arguments):
            raise Exception("Invalid arguments")
        
        # rest of the code goes here

To gracefully catch errors that are generated by functions and class methods, or by the Python interpreter itself, use the `try` and  `except` statements:

    try:
        # normal code goes here
    except SpecialException:
        # code for error handling goes here
        # this code is not executed unless the code
        # above generated a SpecialException error
    except Exception:
        # this code is not executed unless the code 
        # in the try generates an Exception error.
        # the Exception exception is the base class 
        # for all exception objects so should catch 
        # all normal errors.  

For example:

```python
try:
    print(test)
except Exception:
    print("Caught an exception")
```

To get information about the error, we can access the `Exception` class instance that describes the exception by using for example:

    except Exception as e:

```python
try:
    a = 5.0 / 0
except Exception as e:
    print('Caught an exception.')
    raise e
```

We can redirect the code to different places depending on the type of exception:

```python
# Try each of these:

#d = 0
#d = 1
#d = 'a'

try:
    a = 5.0 / d
except ZeroDivisionError as e:
    print('Skipping ZeroDivisionError. {0}'.format(e)) 
except Exception as e:
    msg = "Caught an exception we need to raise."
    raise Exception(msg, e)
a
```

It is easy to make your own exceptions:

```python
class MySpecialException(Exception):
    pass

try:
    a = 5.0 / 'a'
except Exception as e:
    msg = "Something went wrong in MySpecial application."
    raise MySpecialException(msg, e)
```
